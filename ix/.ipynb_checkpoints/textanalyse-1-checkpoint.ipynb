<h1>NLP<h1>

<h2>Vorbereitung</h2>

Nun benötigen wir die spaCY-Bibliothek und da Modell für die deutsche Sprache.

Mehr zu spaCY: https://spacy.io

```python
import spacy
from spacy.lang.de import German

nlp = spacy.load('de')
```

Das durch spaCY erzeugte NLP-Objekt ist sehr groß und benötigt viel Speicher. Um das etwas zu entschärfen,
werden nur nur die benötigen Daten, also Anzahl der Token, Adjektive, Substantive und Verben gespeichert. 
Dabei hilft defaultdict, weil hier alle Werte auf 0 gesetzt sind, das spart viele Fallunterscheidungen.

Die Worte selbst werden auch in ihrer lemmatisierten Form (also grammatikalisch auf die Grundform reduziert)
benötigt, um später Analysen damit durchführen zu können.

```python
from collections import defaultdict

nouns = defaultdict(lambda: 0)
verbs = defaultdict(lambda: 0)
adjectives = defaultdict(lambda: 0)

for i, row in data.iterrows():
  doc = nlp(row["headline"] + " " + row["text"])
  data.set_value(i, "nr_token", len(list(map(lambda x: x.text, 
                                         filter(lambda x: x.pos_ != 'PUNCT', doc)))))
    
  for a in map(lambda x: x.lemma_, filter(lambda x: x.pos_ == 'ADJ', doc)):
    adjectives[a] += 1
  data.set_value(i, "nr_adj", len(list(map(lambda x: x.text, 
                                         filter(lambda x: x.pos_ == 'ADJ', doc)))))  
    
  for n in map(lambda x: x.lemma_, filter(lambda x: x.pos_ == 'NOUN', doc)):
    nouns[n] +=1 
  data.set_value(i, "nr_noun", len(list(map(lambda x: x.text, 
                                         filter(lambda x: x.pos_ == 'NOUN', doc)))))

  for v in map(lambda x: x.lemma_, filter(lambda x: (x.pos_ == 'AUX') | (x.pos_ == 'VERB'), doc)):
    verbs[v] += 1
  data.set_value(i, "nr_verb", len(list(map(lambda x: x.text, 
                                         filter(lambda x: (x.pos_ == 'AUX') | (x.pos_ == 'VERB'), doc)))))        
```

<h2>Keywords über die am ausführlichsten berichtet wird</h2>

Keyword ist ein mehrwertiges Feld, das zunächst mit Hilfe von itertools in ein flaches Array umgewandelt wird.
Mit numpy können anschließend die Häufigkeiten gezählt werden und die häufigsten Keywords mit argsort an das Ende gebracht werden. Interessant sind somit die 50 letzten Einträge.

Mehr zu itertools und chain: https://docs.python.org/3/library/itertools.html

Mehr zu numpy.argsort: https://docs.scipy.org/doc/numpy/reference/generated/numpy.argsort.html

```python
import numpy as np
from itertools import chain, permutations
all_keywords = list(chain(*[l for l in data["keyword"].values if type(l) is list]))
u, indices = np.unique(all_keywords, return_inverse=True)
top_keywords = u[np.argsort(np.bincount(indices))[-50::]].tolist()
top_keyword_counts = np.sort(np.bincount(indices))[-50::].tolist()
```

Für jedes Keyword wird nun mit Hilfe von Pandas die mittlere Artikellänge berechnet. 
Ältere Artikel haben noch keine Keywords. Um das nicht ständig berücksichtigen zu müssen,
trennen wir uns frühzeitig mittels dropna davon uns arbeiten nur noch mit diesen bereinigten
Daten - ein häufiges Data Science-Muster.

Mehr zu apply: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.apply.html

```python
keyword_length = []
data_with_keyword = data.dropna(subset=["keyword"])
for k in top_keywords:
    filter_data = data_with_keyword[data_with_keyword.apply(lambda r: k in r['keyword'], axis=1)]
    keyword_length.append(filter_data["nr_token"].mean())

top_keyword_length = pd.DataFrame()
top_keyword_length["keyword"] = top_keywords
top_keyword_length["#tokens"] = keyword_length
top_keyword_length.sort_values("#tokens", ascending=False).head(10)
```

<h2>Zeitliche Verteilung der Keywords</h2>

Die Berechnung der Monat-Keyword-Matrix nutzt den oben bereits angelegten DataFrame mit den Keywords 
und aggregiert diesen nur auf Monate, dazu gibt es leider keine Resample-Funktion.

Mehr zu apply: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.apply.html

```python
keyword_week = []
for k in top_keywords:
    current = [0 for i in range(0, 12)]
    filter_data = data_with_keyword[data_with_keyword.apply(lambda r: k in r['keyword'], axis=1)]
    total = len(filter_data)
    for ts in filter_data.timestamp:
        current[ts.month-1] += 100.0/total
    keyword_week.append(current)
    
```

Ausreißer sind Keywords, die wirklich nur in einem einzigen Monat vorkommen und leider die Darstellung
der Heatmap ruinieren. Diese werden bei 30% gekappt.

```python
keyword_normalized = [[min([v, 30]) for v in kw] for kw in keyword_week]
```

seaborn baut auf matplotlib auf, kann aber weit komplexere Darstellungen. Für uns ist die
Heatmap sehr nützlich. Die Größe der Grafik wird über plt.figure eingestellt.

Mehr zu Seaborn: https://seaborn.pydata.org

Mehr zur Heatmap: https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
import seaborn as sns
plt.figure(figsize=(12,12))
sns.heatmap(pd.DataFrame(keyword_normalized, columns=range(1,13), index=top_keywords));
```

<h2>Themen der Autoren</h2>

Startpunkt sind die Autoren mit den meisten beiträgen, aus den Keywords wird mit chain für jeden Autor eine flache Liste von Keywords erstellt, die er/sie benutzt hat. Anschließend wird die Auftrittswahrscheinlichkeit für die Top-50-Keywords berechnet.

```python
top_authors = top.author.values
author_keyword = []
for a in top_authors:
    current = []
    author_keywords = list(chain(*[l for l in data.loc[data.author == a].keyword if type(l) is list]))
    for k in top_keywords:
        current.append(author_keywords.count(k)*1.0/(len(author_keywords)+1))
    author_keyword.append(current)
```

Die Darstellung der Heatmap erfolgt genau wie oben, dazu wird wegen der korrekten 
Beschriftung auch wieder alles in einen DataFrame gepackt.

```python
# global setting
plt.figure(figsize=(12,12))
sns.heatmap(pd.DataFrame(author_keyword, columns=top_keywords, index=top_authors));
```

Genau wie in den vorhergenden Keyword-Analysen muss auch hier wieder darauf geachtet

<h2>Graph für Keywords</h2>

Hier werden die oben schon berechneten Top-Keywords, um Artikel auf ihre Verwendung zu überprüfen 
und anschließend die Kanten des Graphen entsprechend zu gewichten. Die Knoten sind die
Keywords selbst.

```python
edges = {}
for kl in data["keyword"].values:
    if isinstance(kl, list):
        for p in permutations(kl, 2):
            if (p[0] in top_keywords) & (p[1] in top_keywords):
                i0 = top_keywords.index(p[0])
                i1 = top_keywords.index(p[1])
                try:
                    edges[(i0, i1)] += 1
                except:
                    edges[(i0, i1)] = 1
```

Dies ist eine sehr krude Art, GraphML auszugeben. Zunächst wird der Header geschrieben,
anschließend alle Nodes (= Keywords) mit Größen, die mit er Wurzel ihrer Häufigkeit 
skaliert werden.

Schließlich sind die Edges an der Reihe. Diese brauchen einen eindeutigen Schlüssel,
der über eine inkrementierte Variable eid realisiert wird.

Die Daten können dann in Gephi importiert werden.

Mehr zu GraphML: https://en.wikipedia.org/wiki/GraphML

Mehr zu Gephi: https://gephi.org

```python
from math import sqrt
max_size = max(top_keyword_counts)
max_weight = max(edges.values())
with open("keyword-network.graphml", "w") as f:
    f.write("""<?xml version="1.0" encoding="UTF-8"?>
<graphml xmlns="http://graphml.graphdrawing.org/xmlns">
<key attr.name="label" attr.type="string" for="node" id="label"/>
<key attr.name="Edge Label" attr.type="string" for="edge" id="edgelabel"/>
<key attr.name="weight" attr.type="double" for="edge" id="weight"/>
<key attr.name="size" attr.type="float" for="node" id="size"/>
<graph edgedefault="undirected">
    """)
    for i, k in enumerate(top_keywords):
        f.write("""<node id="%d">}
  <data key="label">%s</data>
  <data key="size">%f</data>
  </node>
        """ % (i, k, 30*sqrt(top_keyword_counts[i]*1.0/max_size)))
    
    eid = 0
    for st, weight in edges.items():
        f.write("""<edge id="#%d" source="%d" target="%d">
  <data key="weight">%f</data>
  <data key="edgelabel">%d</data>
  </edge>
        """ % (eid, st[0], st[1], 10*sqrt(weight*1.0/max_weight), weight))
        eid += 1
        
    f.write("""</graph>
    </graphml>""")
```

<h2>Wordcloud</h2>

Das WordCloud-Paket muss separat installiert und importiert werden.

Mehr dazu: https://github.com/amueller/word_cloud

```python
from wordcloud import WordCloud
```

Es hat sich gezeigt, dass Word-Clouds mit einem Frequency-Array deutlich schneller und mit
weniger Speicherbedarf erstellt werden können. Daher haben wir ganz oben in der spaCY-Schleife
gleich die entsprechenden Dicts angelegt.

Die erste Darstellung enthält die Hauptworte.

```python
wc = WordCloud(background_color="white", max_words=1000)
wc.generate_from_frequencies(nouns)
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off");
```

Nun sind die Adjektive an der Reihe

```python
wc = WordCloud(background_color="white", max_words=1000)
wc.generate_from_frequencies(adjectives)
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off");
```

Da "neu" und "neue" separat dargestellt werden (Fehler in der Lemmatisierung), werden die Häufigkeiten
addiert und die von "neue" auf 0 gesetzt.

```python
clean = dict(adjectives)
clean["neu"] = clean["neu"] + clean["neue"]
clean["neue"] = 0
wc = WordCloud(background_color="white", max_words=1000)
wc.generate_from_frequencies(clean)
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off");
```
